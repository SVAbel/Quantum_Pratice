```python
qpe = QuantumCircuit(4, 3)
qpe.x(3)
qpe.draw()
```

```python
for qubit in range(3):
    qpe.h(qubit)
qpe.draw()
```

```python
repetitions = 1
for counting_qubit in range(3):
    for i in range(repetitions):
        qpe.cu1(math.pi/4, counting_qubit, 3); # This is C-U
    repetitions *= 2
qpe.draw()
```
